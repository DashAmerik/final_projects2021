```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
data = {'Date': [],
        'Price': [], 
       }
```

```python
i = 0
with open(r'/Users/edwardgaluzo/Downloads/BTC-USD.csv','r') as file:
    reader = csv.reader(file)
    for row in reader:
        if i > 0:
            data['Date'].append(row[0].split("-")[0])#append[row[1]]
            data['Price'].append(int(float(row[1])))
        i+=1
        
```

```python
df = pd.DataFrame(data,columns=['Date','Price'])
df.plot(x ='Date', y='Price', kind = 'line')
plt.show()
```
