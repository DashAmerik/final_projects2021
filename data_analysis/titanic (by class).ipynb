```python
import matplotlib.pyplot as plt
import csv
import numpy as np
sum3 = 0
sum2 = 0
sum1 = 0
alive3 = 0
alive2 = 0
alive1 = 0
```

```python
i = 0
with open(r'C:\Users\DAmer\OneDrive\Документы\GitHub\final_projects2021\data_analysis\Titanic_R.csv','r') as file:
    reader = csv.reader(file, delimiter=';')
    for row in reader:
        if i > 0:
            if row[0] == '1':
                sum1 += 1
                alive1 += int(row[1])
            if row[0] == '2':
                sum2 += 1
                alive2 += int(row[1])
            if row[0] == '3':    
                sum3 += 1
                alive3 += int(row[1])
        i+=1
```

```python
dead1 = sum1 - alive1
dead2 = sum2 - alive2
dead3 = sum3 - alive3
```

```python
y = np.array([alive1, dead1])
mylabels = ["alive (first class)","dead (first class)"]
plt.pie(y, labels = mylabels)
plt.show()

y2 = np.array([alive2, dead2])
mylabels2 = ["alive (second class)","dead (second class)"]
plt.pie(y2, labels = mylabels2)
plt.show()

y3 = np.array([alive3, dead3])
mylabels3 = ["alive (third class)","dead (third class)"]
plt.pie(y3, labels = mylabels3)
plt.show()
```
