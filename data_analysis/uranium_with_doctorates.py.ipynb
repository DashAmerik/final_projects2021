```python
import matplotlib.pyplot as plt
import csv
dates = [1996,1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008]
math_doctorates = [1122,1123,1177,1083,1050,1010,919,993,1076,1205,1325,1393,1399]
uranium_stored = [66.1,65.9,65.8,58.3,54.8,55.6,53.5,45.6,57.7,64.7,77.5,81.2,81.9]
```

```python
x_values = dates
y_values = math_doctorates

second_x_values = dates
second_y_values = uranium_stored


fig = plt.figure()
ax=fig.add_subplot(111)
ax2=fig.add_subplot(111, frame_on=False)

ax.plot(x_values, y_values, color="black")
ax2.plot(second_x_values, second_y_values, color="red")
ax2.xaxis.tick_top()
ax2.yaxis.tick_right()
fig.autofmt_xdate(bottom=0.2, rotation=45, ha='right', which='major')
#plt.xticks(rotation = 45)

plt.show()
```
