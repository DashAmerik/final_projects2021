```python
import matplotlib.pyplot as plt
import csv


x=[]
y=[]

with open(r'C:\Users/DAmer/OneDrive/Документы/GitHub/final_projects2021/data_analysis/example1.csv','r') as file:
    reader = csv.reader(file)
    for row in reader:
        x.append(row[0])
        y.append(int(row[1]))
       
print(x,y)

plt.plot(x,y)


plt.show()
```
